```python
# -------------- Pakete

from pathlib import Path

from imblearn.over_sampling import SMOTE

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier

from minisom import MiniSom

from collections import defaultdict, Counter


# ------------- Importe

import numpy as np
import pandas as pd
```

<div class="alert alert-warning">
    <b>Pfad festlegen</b><br>
    - Pfad zu einer zu verwenden Datenbank festlegen (als .csv)
</div>

```python
# Hauptordner
base_dir = Path.cwd()

# Pfad zur Datenbank
db_root = (
    base_dir.parent
    / "1.2_Rock-Type-Mapping"
    / "Rock-Type_Mapping"
)
```

<div class="alert alert-info">
    <b>Zeitstempel-Ordner</b><br>
    - Falls die Dateien in Zeitstempel Ordner gespeichert worden sind, kann folgender Code verwendet werden
</div>

```python
if not db_root.exists():
    raise FileNotFoundError(f"Pfad existiert nicht: {db_root}")

# Ordner mit neustem Zeitstempel finden
timestamp_folders = [f for f in db_root.iterdir() if f.is_dir()]
if not timestamp_folders:
    raise RuntimeError("Keine Ordner mit Zeitstemepln gefunden")

latest_folder = max(timestamp_folders, key=lambda f: f.stat().st_mtime)
print("Neuester Ordner:", latest_folder.name)

# CSV-Datei darin finden - es sollte nur eine vorhanden sein
csv_files = list(latest_folder.glob("*.csv"))
if not csv_files:
    raise RuntimeError(f"Keine CSV-Datei gefunden: {latest_folder}")

database_path = csv_files[0]
print("Verwendete Datenbank:", database_path.name)

# Datenbank laden
df = pd.read_csv(database_path, low_memory=False)
print("Datenbank geladen. Zeilen:", len(df))
df.head()
```

<div class="alert alert-info">
    <b>Erste Übersicht der Zielvariable "Rock Type"</b><br>
    - Rock-Type wird ale erste Zielvariable verwendet
    - Diese kann in folgendem betrachtet werden
</div>

```python
print(df.shape) # Zeilen, Spalten
print(df.columns) # Spaltennamen
df.info() # Datentypen
df.describe().T.head(30) # statistische Übersicht für numerische Spalten
```

```python
# Testen ob Zielvariable "rock_type" existiert
if "rock_type" not in df.columns:
    raise KeyError("'rock_type' fehlt in der Datenbank")

# Überprüfen
df["rock_type"].value_counts(dropna=False).head(20)
```

<div class="alert alert-info">
    <b>Preprocessing</b><br>
    - Feature Festlegung - Dies kann basierend auf der vorhergehenden explorativen Datenanalyse geschehen
</div>

```python
feature_cols = [
    "HCO3_in_mmol/L",
    "SO4_in_mmol/L",
    "pH",
    "Na_in_mmol/L",
    "Mg_in_mmol/L",
    "Ca_in_mmol/L",
    "Cl_in_mmol/L",
    "total_dissolved_solids_in_mmol/L"
]
```

```python
target_col = "rock_type"

# Prüfen, welche Spalten passen
existing_features = [c for c in feature_cols if c in df.columns]
missing_features = [c for c in feature_cols if c not in df.columns]

print("Vorhandene Feature-Spalten:", existing_features)
print("Nicht vorhandene Feature-Spalten:", missing_features)

# Nur wenn gültiger Rock Type
df_valid = df.dropna(subset=[target_col]).copy()

# Feature Spalten in numerische Typen umwandeln
for col in feature_cols:
    df_valid[col] = pd.to_numeric(df_valid[col], errors="coerce")

# Fehlende Werte pro Feature
missing_counts = df_valid[existing_features].isna().sum()
print("\nFehlende Werte pro Feature:\n", missing_counts)

# Zählen aller Zeilen mit vollständigen Features:
df_full = df_valid.dropna(subset=feature_cols)

print("\nGesamtzeilen mit rock_type:", len(df_valid))
print("Zeilen mit vollständigen Features:", len(df_full))
print("Verlust an Zeilen:", len(df_valid) - len(df_full))

# Optional: erste Zeilen ausgeben
df_full.head()
```

```python
X = df_full[feature_cols]
y = df_full[target_col]

# Labels encoden
encoder = LabelEncoder()
y_enc = encoder.fit_transform(y)

# Train/Test-Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=0.2, random_state=42, stratify=y_enc
)

# Pipeline: Skalierung + SVC mit Klassen-Gewichtung
clf = Pipeline([
    ("scaler", StandardScaler()),
    ("svc", SVC(
        kernel="rbf",
        C=10,
        gamma="scale",
        class_weight="balanced"   # WICHTIG: korrigiert starke Unbalance
    ))
])

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred, target_names=encoder.classes_))

print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, y_pred))
```

<div class="alert alert-danger">
    <b>Mapping auf vereinfachte Klassen</b><br>
    Muss noch bearbeitet und wissenschaftlich fundierter sein

```python
mapping = {
    # SEDIMENTARY
    "siliciclastic sedimentary rocks": "SEDIMENTARY",
    "mixed sedimentary rocks": "SEDIMENTARY",
    "carbonate sedimentary rocks": "SEDIMENTARY",
    "evaporites": "SEDIMENTARY",
    "unconsolidated sediments": "SEDIMENTARY",

    # IGNEOUS (MAGMATIC)
    "acid plutonic rocks": "IGNEOUS",
    "intermediate plutonic rocks": "IGNEOUS",
    "basic plutonic rocks": "IGNEOUS",
    "acid volcanic rocks": "IGNEOUS",
    "intermediate volcanic rocks": "IGNEOUS",
    "basic volcanic rocks": "IGNEOUS",
    "pyroclastic rocks": "IGNEOUS",

    # METAMORPHIC
    "metamorphic rocks": "METAMORPHIC",

    # WATER - als non-rock klassifizieren
    "water bodies": "NON_ROCK"
}

print(df["rock_type_simplified"].value_counts(dropna=False))
df["rock_type_simplified"] = df["rock_type"].map(mapping)
```

<div class="alert alert-warning">
    <b>Bereinigung, da im vorherigen Schritt noch fehler auftreten</b><br>
    - Diese Fehler müssen im Nachgang noch bearbeitet werden<br>
    - Aktuell NaN und Non-Rocks entfernen
</div>

```python
# Zeilen mit gültiger vereinfachter Klasse
df_simpl = df.dropna(subset=["rock_type_simplified"]).copy()

# Non-Rocls entfernen
df_simpl = df_simpl[df_simpl["rock_type_simplified"] != "NON_ROCK"]

print("Zeilen nach Entfernen von NaN und Non-Rock:", len(df_simpl))

# Feature-Spalten numerisch
for col in feature_cols:
    df_simpl[col] = pd.to_numeric(df_simpl[col], errors="coerce")

# NaN entfernen
df_simpl_clean = df_simpl.dropna(subset=feature_cols)

print("Zeilen nach Feature-Reinigung:", len(df_simpl_clean))

# Ausgabe der neuen Klassenzahlen
print("\nNeue Klassenverteilung:")
print(df_simpl_clean["rock_type_simplified"].value_counts())
```

```python
df_simple = df.dropna(subset=["rock_type_simplified"]).copy()
```

```python
target_col = "rock_type_simplified"
```

<div class="alert alert-danger">
    <b>Versuch SOM</b><br>
    - Muss vsl. noch geändert werden<br>
    - soll eine erste Version des Trainings für SOM darstellen
</div>

```python
# Features und Ziel setzen
X = df_simpl_clean[feature_cols].values
y = df_simpl_clean["rock_type_simplified"].values

# Label-Encoding der Klassen
encoder = LabelEncoder()
y_enc = encoder.fit_transform(y)

# Train-Test-Split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_enc,
    test_size=0.2,
    random_state=42,
    stratify=y_enc
)

# Skalierung der Eingabe
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Trainingsdaten-Shape für SOM:", X_train_scaled.shape)

# SOM konfigurieren und trainieren
som_x, som_y = 20, 20

som = MiniSom(
    x=som_x,
    y=som_y,
    input_len=X_train_scaled.shape[1],
    sigma=1.0,
    learning_rate=0.5,
    neighborhood_function='gaussian',
    random_seed=42,
)

som.random_weights_init(X_train_scaled)
print("SOM-Training starten")
som.train_random(X_train_scaled, num_iteration=5000)
print("SOM-Training abgeschlossen")

# Neuron-zu-Klassen-Mapping auf Basis der Trainingsdaten
bmu_label_map = defaultdict(list)

for xi, label in zip(X_train_scaled, y_train):
    winner = som.winner(xi)  # (i, j) Koordinate des Bestens Matchings
    bmu_label_map[winner].append(label)

# Für jedes Neuron die häufigste Klasse bestimmen
bmu_class_map = {}
for bmu, labels in bmu_label_map.items():
    most_common_label = Counter(labels).most_common(1)[0][0]
    bmu_class_map[bmu] = most_common_label

# Vorhersagen für Testdaten
y_pred = []
global_majority = Counter(y_train).most_common(1)[0][0]

for xi in X_test_scaled:
    winner = som.winner(xi)
    if winner in bmu_class_map:
        y_pred.append(bmu_class_map[winner])
    else:
        # Fallback: globale Mehrheitsklasse aus dem Training
        y_pred.append(global_majority)

y_pred = np.array(y_pred)

# Auswertung
print("\nClassification Report (SOM mit Neuron-Labeling):\n")
print(classification_report(y_test, y_pred, target_names=encoder.classes_))

print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, y_pred))
```
