<div class="alert alert-block alert-warning">
    https://doi.pangaea.de/10.1594/PANGAEA.981233<br><br>
    DATENBANK 3: PANGEA Hotsprings USA
</div>

```python
# -------------------------- Pfad ---------------------------
dateipfad = pfad / "NOAA_NCEI_US_hot_springs_1978-1985.tab"

# -------------------------- Kommentarende suchen ------------
skip = 0
with open(dateipfad, encoding="latin1") as f:
    for i, line in enumerate(f):
        if line.strip().startswith("*/"):
            skip = i + 1   # +1, damit wir ab der nächsten Zeile (also nach */) starten
            break

print(f"→ Überspringe {skip} Kommentarzeilen (bis einschließlich */)")


# -------------------------- Einlesen ------------------------
df = pd.read_csv(
    dateipfad,
    sep="\t",
    encoding="latin1",
    skiprows=skip,
    engine="python"
)


# -------------------------- Ausgabe -------------------------
display(df.head(10))
print(f"\nZeilen: {len(df)},  Spalten: {len(df.columns)}")
```
